```python
#Part 1
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report

# Load the Iris dataset
iris_data_path = 'Iris - Iris.csv'
iris_df = pd.read_csv(iris_data_path)

# Drop the 'Id' column as it's not a feature for model training
iris_df.drop('Id', axis=1, inplace=True)

# Separate the dataset into features (X) and the target label (y)
X = iris_df.drop('Species', axis=1)
y = iris_df['Species']

# Split the dataset into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the model on the training set
clf.fit(X_train, y_train)

# Predict the species for the test set
y_pred = clf.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

# Calculate the precision of the model
precision = precision_score(y_test, y_pred, average='macro')

# Calculate the recall of the model
recall = recall_score(y_test, y_pred, average='macro')

# Generate a classification report
report = classification_report(y_test, y_pred)

# Print the model evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Classification Report:\n", report)
```

```python
# Part 2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris_data_path = 'Iris - Iris.csv'
iris_df = pd.read_csv(iris_data_path)

# Display statistical summaries of the features
print(iris_df.describe())

# Display information about the DataFrame including the data types and memory usage
print(iris_df.info())

# List of feature names for easier access
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Plot histograms for each feature to understand the distribution
iris_df[features].hist(bins=15, figsize=(10, 6))
plt.show()

# Use seaborn's pairplot to visualize pairwise relationships in the dataset
# The different species are colored differently (hue='Species')
sns.pairplot(iris_df, hue='Species', markers=["o", "s", "D"])
plt.show()

# Features to be used for box plots
numeric_features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Plot box plots for each feature, categorized by species
# This helps to visually compare the distributions and spot any potential outliers
for feature in numeric_features:
    iris_df.boxplot(column=feature, by='Species', figsize=(8, 5))
    plt.show()
```
